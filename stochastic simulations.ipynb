```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# here we want to simulate random walks
N = 1000 # how many trials
L = 1000 # length of run
dt = .1
D = 1
t = np.arange(L+1)*dt

dW = np.random.normal(loc=0.0, scale=np.sqrt(2*D*dt), size=(N,L))

# starting values are all zero
start = np.zeros(N)

# add starting values to displacements and calculate path
x = np.column_stack((start, dW))
paths = np.cumsum(x, axis=1)

for path in paths:
    plt.plot(t,path)
```

```python
# lets see whether we can analyze the data
# we know that <x^2> should increase as 2Dt
x_square = np.std(paths,axis=0)**2
plt.plot(t,x_square)
```

```python
# we can now fit a line to this curve
p = np.polyfit(t,x_square,1)
fitp = np.poly1d(p)
print(p)
plt.plot(t,x_square)
plt.plot(t,fitp(t),label="slope: "+str(p[0]))
plt.legend()
```
