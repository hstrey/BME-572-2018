```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
egfp=np.array(pd.read_excel('20170316_K1_gfp_timecorr.xlsx', index_col=None,header=None)).T
time = np.array(egfp[0]/60/60)
egfp = egfp[1:]
print(egfp.shape)
```

```python
for i,g in enumerate(egfp):
    plt.figure()
    plt.plot(time,g,label=str(i))
    plt.legend()
```
